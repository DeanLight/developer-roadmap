## generate content stubs

```python
balsamiq_json_path = Path('public/project/ai.json')

content_dir_base_path = Path('content/roadmaps/112-ai/content/')
content_dir_base_path.mkdir(parents=True, exist_ok=True)

content_dir_path_relative = Path('/')/content_dir_base_path.relative_to('content')

content_paths_json_path = Path('content/roadmaps/112-ai/content-paths.json')
```

```python
balsamiq_json_path = Path('public/project/ai.json')

with open(balsamiq_json_path) as f:
    balsamiq_json = json.load(f)

node_names = jq('.. | objects | with_entries(select(.key | contains("controlName"))) | select(. != {}) | .controlName ').input(balsamiq_json).all()
#node_names= node_names[-10:]
node_names
```

```python
paths_jsons = {"home": str(content_dir_path_relative / 'readme.md')}

for node_name in node_names:
    name_parts = node_name.split(':')
    dir_name = name_parts[0]
    if len(name_parts) == 1:
        file_name = 'readme'
    else:
        file_name = name_parts[-1]
    prefix_less_name = node_name[4:]
    paths_jsons[prefix_less_name] = str(content_dir_path_relative / dir_name / f'{file_name}.md')

paths_jsons
```

```python
with open(content_paths_json_path, 'w') as f:
    json.dump(paths_jsons, f, indent=2)
```

```python
def create_stubs(paths_json,dryrun=False):
    for node_name, path in paths_jsons.items():
        content_path = Path('content')/Path(path).relative_to('/')
        if content_path.exists():
            print(f"{content_path} already exists")
        else:
            if dryrun:
                print(f"Creating {content_path}")
            else:
                content_path.parent.mkdir(parents=True, exist_ok=True)
                with open(content_path, 'w') as f:
                    print(f"writing stub to {content_path}")
                    f.write(f'# {node_name}')

create_stubs(paths_jsons,dryrun=False)
```
